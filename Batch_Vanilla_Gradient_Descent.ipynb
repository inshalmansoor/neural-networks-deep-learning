```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple dataset (y = 2x + 1 with noise)
np.random.seed(42)
X = np.linspace(0, 10, 50)
true_m, true_c = 2, 1
y = true_m * X + true_c + np.random.normal(0, 2, X.shape)  # Noise
```

```python
# Initialize parameters (slope and intercept)
m, c = 0.0, 0.0
learning_rate = 0.01
iterations = 1000
```

```python
loss_history = []
```

```python
# Gradient Descent
for i in range(iterations):
    y_pred = m * X + c

    m_grad = (-2 / len(X)) * np.sum(X * (y - y_pred))
    c_grad = (-2 / len(X)) * np.sum(y - y_pred)

    m = m - learning_rate * m_grad
    c = c - learning_rate * c_grad

    # loss (MSE)
    loss = np.mean((y - y_pred) ** 2)
    loss_history.append(loss)

    if i % 100 == 0:
        print(f"Iteration {i}: Loss = {loss:.4f}, m = {m:.4f}, c = {c:.4f}")
```

```python
print(f"Final parameters: m = {m:.4f}, c = {c:.4f}")
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Points", color="blue")
plt.plot(X, m * X + c, label=f"Fitted Line: y = {m:.2f}x + {c:.2f}", color="red")
plt.title("Gradient Descent: Line Fitting")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Plot loss history
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), loss_history, label="Loss")
plt.title("Loss Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()
```
